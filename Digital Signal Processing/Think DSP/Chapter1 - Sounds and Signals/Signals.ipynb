```python
from thinkdsp import CosSignal, SinSignal, decorate
import matplotlib.pyplot as plt

cosine_signal = CosSignal(freq=440, amp=1.0, offset=0)

sine_signal = SinSignal(freq=880, amp=0.5, offset=0)

cosine_signal.plot()
decorate(xlabel='Time (s)')

plt.show()
```

```python
sine_signal.plot()
decorate(xlabel='Time (s)')

plt.show()
```
