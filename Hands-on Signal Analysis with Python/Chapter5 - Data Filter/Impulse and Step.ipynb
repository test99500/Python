```python
import numpy as np
import matplotlib.pyplot as plt

# Definition of "Impulse"
impulse = np.zeros(50)
impulse[10] = 1

# Definition of "Step"
step = np.zeros(50)
step[10:] = 1

# Show impulse and step
figure, axes = plt.subplots(1, 2)
axes[0].plot(impulse)
axes[0].set_title("Impulse")

axes[1].plot(step)
axes[1].set_title("Step")

plt.show()
```
